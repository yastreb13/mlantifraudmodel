```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\Users\yanen\OneDrive\Рабочий стол\creditcard.csv')
df.head()
```

```python
print(df.isnull().sum())
```

```python
fraud_len=len(df[df['Class']==1])
legit_len=len(df)-fraud_len
print("Количество мошенников: {:10}\t Количество обычных людей: {:10}\t Отношение мошенников ко все длине датафрейма {:10.2f}%".format(fraud_len,legit_len,fraud_len*100/len(df)))
```

```python

class_corr_df=df.corr()[['Class']].transpose()

plt.figure(figsize=(12, 2))
sns.heatmap(class_corr_df, annot=True)
plt.title('Корреляция всех признаков с признаком, указывающим на мошенников')
plt.show()
```

```python
sns.barplot(data=df.corr()[['Class']].drop('Class').transpose())
plt.show()
class_corr=df.corr()['Class'].drop('Class')

print(f'Максимальная: {class_corr.idxmax()} = {class_corr[class_corr.idxmax()]:.3f}')
print(f'Минимальная: {class_corr.idxmin()} = {class_corr[class_corr.idxmin()]:.3f}')
```

```python
df['Hour']=df['Time']//3600%24
legit=df[df['Class']==0]
fraud=df[df['Class']==1]

print("Основные статистики Amount:")
print("Легитимные: mean={:.2f}, median={:.2f}".format(legit['Amount'].mean(),legit['Amount'].median()))
print("Мошенничество: mean={:.2f}, median={:.2f}".format(fraud['Amount'].mean(),fraud['Amount'].median()))

fig, (ax1, ax3, ax2)=plt.subplots(1, 3,  figsize=(14, 5))

ax1.boxplot([legit['Amount'], fraud['Amount']], labels=['Legit', 'Fraud'])
ax1.set_yscale('log')
ax1.set_title('Amount распределение (log scale)')

ax3.boxplot([legit['Amount'], fraud['Amount']], labels=['Legit', 'Fraud'])
ax3.set_title('Amount распределение')

ax2.plot([0.5, 0.9, 0.95, 0.99], legit['Amount'].quantile([0.5, 0.9, 0.95, 0.99]), 
         'o-', label='Legit', markersize=8)
ax2.plot([0.5, 0.9, 0.95, 0.99], fraud['Amount'].quantile([0.5, 0.9, 0.95, 0.99]), 
         's-', label='Fraud', color='red', markersize=8)
ax2.set_title('Квантильное сравнение')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python


fraud_by_hour_percent=df.groupby('Hour')['Class'].mean()*100

plt.figure(figsize=(12, 6))
plt.bar(fraud_by_hour_percent.index, fraud_by_hour_percent.values)
plt.title('Процент мошенничества, в соотвествие с часами')
plt.xlabel('Час дня')
plt.ylabel('% Fraud')
plt.show()

print("Часы с наибольшим количеством Fraud:")
for hour, rate in fraud_by_hour_percent.sort_values(ascending=False).head(3).items():
    print("{:.0f}:00 - {}% мошенничества".format(hour,rate))
```

```python
fig, ax=plt.subplots(figsize=(10, 4))


sample_legit=legit.sample(min(3000, len(legit)), random_state=42)
ax.scatter(sample_legit['Hour'], sample_legit['Amount'], 
           alpha=0.1, s=10, color='blue', label='Legit (sample)')
ax.scatter(fraud['Hour'], fraud['Amount'], 
           alpha=0.7, s=30, color='red', label='Fraud')

ax.set_xlabel('Hour')
ax.set_ylabel('Amount')
ax.set_title('Fraud и Legit по часам')
ax.set_xticks(range(0, 24, 3))
ax.legend()
ax.grid(True, alpha=0.9)
plt.show()

print("Fraud rate по часам:")
Fraud_rate=[]
for hour in range(0,24):
    rate = df[df['Hour'] == hour]['Class'].mean() * 100
    Fraud_rate.append(rate)
    print(f"  {hour:02d}:00 - {rate:.3f}%")
plt.figure(figsize=(10, 4))
sns.lineplot(x=range(24), y=Fraud_rate, color='red')

plt.title('Fraud rate по часам')
plt.xlabel('Hour')
plt.ylabel('Fraud rate (%)')
plt.xticks(range(0, 24))
plt.grid(True, alpha=0.3)
plt.show()
```

```python

fraud_amounts = df.loc[df["Class"] == 1, "Amount"]

cost_fn_median = fraud_amounts.quantile(0.50)

cost_fp_1pct_median = cost_fn_median * 0.01

print("\nЗначения для подсчета потерь")
print(f"cost_fn={cost_fn_median:.2f}, испрользуя медиану")
print(f"cost_fp={cost_fp_1pct_median:.2f}  1% от FN ")
```
